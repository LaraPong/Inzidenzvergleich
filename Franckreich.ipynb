```python
#Auvergne Rhone Alpes
import requests,json
url ="https://dashboardcovid.trial.opendatasoft.com/api/records/1.0/search/?dataset=indicateurs-de-suivi-de-lepidemie-de-covid-19-en-france&q=&sort=extract_date&facet=extract_date&facet=reg_name&facet=dep_name&facet=dep_area_code&refine.reg_name=Auvergne+Rh%C3%B4ne+Alpes&refine.extract_date=2021"
result = requests.get(url=url) #Anfrage absetzen
resultjson = json.loads(result.text)
newjson=resultjson['records'][0]['fields']

for key in newjson:
    print(key, ":", newjson[key])
```
